**Remove Unwanted Columns and Rename the Others**

```python
df = df.drop(['Market Category', 'Vehicle Style',
              'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head()
```

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders",
                        "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode",
                        "highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price", "Engine Fuel Type":"Fuel Type"})
df.head()
```

**Handling Missing Values**

```python
##check for missing values in each column
df.isnull().sum()
```

```python
#check for percentage of missing values in each column
(df.isnull().mean() * 100)
```

```python
#remove missing values
df = df.dropna(subset=['Fuel Type'])
df.count()
```

```python
#Impute missing value with mean
df['HP'] = df['HP'].fillna(df['HP'].mean())
```

```python
#Impute missing value with median
df['Cylinders'] = df['Cylinders'].fillna(df['Cylinders'].median())
```

```python
df.isnull().sum()
```

**Check for Duplicates and remove them**

```python
#check for number of duplicates
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
#remove the duplicates
df = df.drop_duplicates()
df.head()
```

```python
df.count()
```

**Handling Outliers**

```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

# Filter outliers
outliers = df[(df['Price'] < Q1 - 1.5 * IQR) | (df['Price'] > Q3 + 1.5 * IQR)]
outliers
```

```python
Q1 = df['Cylinders'].quantile(0.25)
Q3 = df['Cylinders'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df = df[(df['Cylinders'] >= lower_bound) & (df['Cylinders'] <= upper_bound)]
```

```python
sns.boxplot(x=df['Cylinders'])
```

**Univariate Analysis**

```python
#Histogram: Numerical Variables
sns.histplot(df['HP'], kde=True, color='green')
plt.title("HP Distribution")
plt.xlabel("HP")
plt.ylabel("Count")
plt.show()
```

```python
#Countplot: Categorical Variables
top_10 = df['Make'].value_counts().nlargest(10).index

# Filter the DataFrame to only include those top 10
filtered_data = df[df['Make'].isin(top_10)]


sns.countplot(y='Make', data=filtered_data, order=top_10)
plt.title("Top 10 Most Frequent Categories")
plt.xlabel("No of Cars Sold")
plt.ylabel("Make")
plt.show()
```

**Bivariate Analysis**

```python
#Barplot: comparing between categorical variables

top5 = df['Model'].value_counts().nlargest(5).reset_index()
top5.columns = ['Model', 'MPG-H']


sns.barplot(x='Model', y='MPG-H', data=top5, palette='viridis')
plt.title("Top 5 Model by MPG-H")
plt.xlabel("Model")
plt.ylabel("MPG-H")
plt.show()
```

```python
#Scatterplot: Comparing between numerical variables
sns.scatterplot(x='HP', y='Price', data=df)
plt.title("HP vs Price")
plt.xlabel("HP")
plt.ylabel("Price")
plt.show()
```

```python
#Lineplot: time series comparison
sns.lineplot(x='Year', y='Price', data=df)
plt.title("Price Over Time")
plt.xlabel("Year")
plt.ylabel("Price")
plt.show()
```
